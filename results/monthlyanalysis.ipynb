```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Prepare the data
data = [
    [2010, 12, 43419, 14], [2011, 1, 469795, 144], [2011, 2, 466307, 144],
    [2011, 3, 485165, 150], [2011, 4, 502042, 157], [2011, 5, 561647, 174],
    [2011, 6, 737793, 230], [2011, 7, 596710, 188], [2011, 8, 614516, 193],
    [2011, 9, 603047, 185], [2011, 10, 708164, 221], [2011, 11, 660507, 208],
    [2011, 12, 669395, 222], [2012, 1, 495363, 252], [2012, 2, 506992, 260],
    [2012, 3, 373478, 212], [2012, 4, 400324, 219], [2012, 5, 358866, 207],
    [2012, 6, 555142, 318], [2012, 7, 444533, 246], [2012, 8, 523887, 294],
    [2012, 9, 486149, 269], [2012, 10, 535125, 313], [2012, 11, 537918, 324],
    [2012, 12, 624454, 483], [2013, 1, 857758, 1677], [2013, 2, 771218, 3454],
    [2013, 3, 1049732, 4087], [2013, 4, 1045860, 3979], [2013, 5, 1284456, 4400],
    [2013, 6, 1642948, 5025], [2013, 7, 1371595, 4673], [2013, 8, 1545910, 4848],
    [2013, 9, 1447324, 4616], [2013, 10, 1673261, 5304], [2013, 11, 1780688, 5224],
    [2013, 12, 1874128, 5520], [2014, 1, 45642, 1970]
]

# 2. Create DataFrame and format dates
df = pd.DataFrame(data, columns=['year', 'month', 'revenue', 'quantity'])
df['date'] = pd.to_datetime(df[['year', 'month']].assign(day=1))
df = df.sort_values('date')

# 3. Plotting
fig, ax1 = plt.subplots(figsize=(12, 6))

# Primary Y-axis: Revenue
ax1.set_xlabel('Date')
ax1.set_ylabel('Revenue ($)', color='blue')
ax1.plot(df['date'], df['revenue'], color='blue', marker='o', label='Revenue')
ax1.tick_params(axis='y', labelcolor='blue')

# Secondary Y-axis: Quantity
ax2 = ax1.twinx()
ax2.set_ylabel('Quantity Sold', color='red')
ax2.plot(df['date'], df['quantity'], color='red', marker='x', linestyle='--', label='Quantity')
ax2.tick_params(axis='y', labelcolor='red')

plt.title('Monthly Sales Trend: 2010 - 2014')
plt.grid(True, alpha=0.3)
plt.show()
```
